### Concat(Birleştirme) İşlemi

```python
npArray=np.random.randint(1,15,size=(5,3))
npArray2=np.random.randint(1,15,size=(5,3))

df7=pd.DataFrame(npArray,columns=["Sutun1","Sutun2","Sutun3"])
df8=pd.DataFrame(npArray2,columns=["Sutun1","Sutun2","Sutun3"])
```

```python
df7
```

```python
df8
```

```python
pd.concat([df7,df8])
```

```python
pd.concat([df7,df8],ignore_index=True)
```

#### DİKKAT
Sütunlar uyuşmadığı durumda NaN değerler ile karşılaşabiliriz.

```python
df8["Sutun4"]=[0,1,2,3,4]
pd.concat([df7,df8],ignore_index=True)
```

```python
pd.concat([df7,df8],ignore_index=True,join="inner")
```

### Merge(Birleştirme) İşlemi

```python
import pandas as pd
df_dersler=pd.DataFrame({'ogrenciler':['Ogun','Ogun','Elzem','Beril','Elif','Burak'],'dersler':['Turkce','Matematik','Matematik','Turkce','Fizik','Kimya']})
df_notlar=pd.DataFrame({'ogrenciler':['Ogun','Ogun','Elzem','Beril','Elif','Burak'],'notlar':[100,50,75,95,85,70]})
```

```python
df_sonhali=pd.merge(df_dersler,df_notlar,on="ogrenciler")
df_sonhali
```

```python
df_sonhali=pd.merge(df_dersler,df_notlar,on="ogrenciler")
df_sonhali
```

### Kümeleme ve Gruplama(Aggregation-Group) İşlemleri

Elimizdeki veri setini groupby() fonksiyonu ile belirli sütunlara göre gruplayabiliriz.

Elimizdeki veri setini grupladıktan sonra aggregate() fonksiyonu ile istatiksel hesaplamalar yapabiliriz.

```python
df9.describe()
```

```python
df9.groupby("method").describe().T
```

```python
df9.groupby("method")["orbital_period"].mean()
```

```python
df9.groupby("method")["orbital_period"].aggregate([min,np.mean,max])
```

### Apply Fonksiyonu

```python
dfFiyat=pd.DataFrame([100,150,200,250,300,350],columns=["Fiyat"])
dfFiyat
```

```python
def indirimYap(fiyat):
    return fiyat-fiyat*0.05
```

```python
dfFiyat["Fiyat"].apply(indirimYap)
```

```python
dfFiyat["Fiyat"].apply(lambda fiyat:fiyat-fiyat*0.05)
```
